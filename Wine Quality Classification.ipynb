```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['quality'])
y = df['quality']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=46)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
classifier = LogisticRegression(random_state=42)
classifier.fit(X_train, y_train)
y_pred_log_reg = classifier.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_log_reg))



```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred_tree = decision_tree.predict(X_test)

print("\nDecision Tree Accuracy:", accuracy_score(y_test, y_pred_tree))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_tree))
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)


y_pred_forest = random_forest.predict(X_test)

print("\nRandom Forest Accuracy:", accuracy_score(y_test, y_pred_forest))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_forest))
```

we can clear see that the accuraccy_score of random forest is much better as compared to Logistic regression and decision tree , so i'm going to apply some more algo or tools to boost accuracy score of are model through hyperparameter tunning , so i apply gridseachcv on my model which are build by the help of random forest

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}

CV_random_f = GridSearchCV(estimator=random_forest, param_grid=param_grid, cv= 5)
CV_random_f.fit(X_train, y_train)

```

```python
CV_random_f.best_params_
```

```python
randomfc1=RandomForestClassifier(random_state=42, max_features='sqrt',
                                 n_estimators= 500, max_depth=8, criterion='gini')
```

```python
randomfc1.fit(X_train, y_train)
pred=randomfc1.predict(X_test)
print("Accuracy for Random Forest on CV data: ",accuracy_score(y_test,pred))
```
